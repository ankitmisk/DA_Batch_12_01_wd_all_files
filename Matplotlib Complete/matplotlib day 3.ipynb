```python
# np.random.seed(10)
y1 = np.random.randint(-100,1000,10)
y2 = np.random.randint(-100,1000,10)
y3 = np.random.randint(-100,1000,10)



#'facecolor':'r'
ax = plt.boxplot([y1,y2,y3],
                 notch = True,patch_artist=True,showcaps=True,showbox=True,
                 showmeans=True,showfliers=True,boxprops = {'color':'r','linestyle':'.-','linewidth':1},
                 flierprops={'color':'r'})



all_colors = ['yellow','g','b']


for i,j in enumerate(ax['boxes']):
    j.set_facecolor(all_colors[i])

plt.show()
```

```python
import calendar
days = list(calendar.day_name)
sales = np.random.randint(1,1000,7)

plt.figure(figsize=(7,3))

ax = plt.plot(days,sales)
max_sales_idx = np.argmax(np.array(sales))

max_day = days[max_sales_idx]

plt.xticks(rotation = 45)


min_day = days[np.argmin(np.array(sales))]

print(min_day)

plt.annotate(text = 'Max Tips',xy = (max_day,max(sales)),xytext=(min_day,max(sales)-20),arrowprops = {'arrowstyle':"->"})
plt.show()
```

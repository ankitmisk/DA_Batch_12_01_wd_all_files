```python
x = ['Python','Java','C','C++','Php']
y = np.random.randint(1,50,5)


ax = plt.pie(y,labels=x,
             colors=sns.color_palette('magma'),
             labeldistance=0.7,textprops={'color':'w'},
            autopct = '%.2f%%',pctdistance = 0.5)
plt.show()
```

```python
np.random.seed(5)

x = ['Python','Java','C','C++','Php']
y = np.random.randint(1,50,5)

exp = list(np.array([0]*5))
exp[np.argmin(y)] = 0.2


ax = plt.pie(y,labels=x,
             colors=sns.color_palette('hot'),
             labeldistance=1.5,textprops={'color':'k'},
            autopct = '%.2f%%',pctdistance = 1.25,
            explode = exp,hatch = '..')
plt.show()
```

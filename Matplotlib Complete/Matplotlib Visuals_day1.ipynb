```python
from matplotlib import pyplot as plt
```

```python
x = list('abcde'.upper())
y1 = np.random.randint(1,100,5)
y2 = np.random.randint(1,100,5)

plt.figure(figsize=(7,3))

plt.plot(x,y1,y2,linestyle = '-.',alpha = 0.5,label = ['Estimated Marks','Actual marks'])
# plt.plot(x,y2,alpha = 0.5)
plt.xlabel('Class',{'fontsize':15,'color':'g'})
plt.ylabel('Avg Marks',fontsize = 15,color = 'r')
plt.title('Score Card'.upper(),fontsize = 15,color = 'r')
plt.grid(alpha= 0.3)

plt.legend()
plt.show()
```

```python
x = list('abcde'.upper())
# np.random.seed(1)

y1 = np.random.randint(1,100,5)
y2 = np.random.randint(1,100,5)

plt.figure(figsize=(7,3))

plt.plot(x,y1,linestyle = '-.',alpha = 0.5,label = 'Estimated Marks')
plt.plot(x,y2,alpha = 0.5,label = 'Actual Marks',marker = 'o')

plt.xlabel('Class',{'fontsize':15,'color':'g'})
plt.ylabel('Avg Marks',fontsize = 15,color = 'r')
plt.title('Score Card'.upper(),fontsize = 15,color = 'r')
plt.grid(alpha= 0.3)

for i, j in zip(x,y1):
    plt.text(i,j-3,str(j))

    
plt.xticks(color='r')    
plt.yticks(fontsize=12)
plt.legend()
plt.savefig('class report.png',dpi=1000)
plt.show()
```

```python

for k in range(1,11):

    x = list('abcde'.upper())
    # np.random.seed(1)

    y1 = np.random.randint(1,100,5)
    y2 = np.random.randint(1,100,5)

    plt.figure(figsize=(7,3))

    plt.plot(x,y1,linestyle = '-.',alpha = 0.5,label = 'Estimated Marks')
    plt.plot(x,y2,alpha = 0.5,label = 'Actual Marks',marker = 'o')

    plt.xlabel('Class',{'fontsize':15,'color':'g'})
    plt.ylabel('Avg Marks',fontsize = 15,color = 'r')
    plt.title(f'Score Card {k}'.upper(),fontsize = 15,color = 'r')
    plt.grid(alpha= 0.3)

    for i, j in zip(x,y1):
        plt.text(i,j-3,str(j))


    plt.xticks(color='r')    
    plt.yticks(fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.savefig(path+f'\\image_{k}.png',dpi=1000)
    plt.show()
    display(clear=True)
```

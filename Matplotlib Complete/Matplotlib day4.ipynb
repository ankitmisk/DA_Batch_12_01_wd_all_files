```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.linspace(-5,5,50)
x = range(0,len(y))
y = np.sin(y)
ax = plt.stem(x,y,linefmt= 'red',markerfmt='x')
plt.plot(x,y)
plt.show()
```

```python
fig,ax = plt.subplots(2,2)

y = np.linspace(-5,5,50)
y = np.sin(y)
x = range(len(y))
# plt.subplots_adjust(bottom=0.01, right=0.8, top=1)

data = np.random.randint(1000,5000,5)
lab = 'ABCDE'

fig.supxlabel('X label')
fig.supylabel('Y label')
fig.suptitle('Multiple Subplots')

y_1 = np.random.normal(size = 100)
y_2 = np.random.normal(size = 100)
y_3 = np.random.normal(size = 100)

# ax[0,1].set_alpha(0)
ax[0,0].plot(x,y)
ax[0,1].pie(data,labels = lab)
ax[1,0].bar(list(lab),height = data)
ax[1,1].boxplot([y_1,y_2,y_3])
# ax[0,0].set_xlabel('Sin Curve')
plt.show()
```

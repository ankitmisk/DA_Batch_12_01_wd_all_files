```python
text = np.array(list('ABC')*3).reshape(3,3).astype('<U32')
```

```python
text
```

```python
corr_text = tips.corr(numeric_only=True).to_numpy().round(2).astype('<U32')
corr_text
```

```python
final_text = []
for i,j in zip(text.flatten(),corr_text.flatten()):
    final_text.append(i+'\n'+j)
    
fiinal_text = np.array(final_text).reshape(3,3)
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=fiinal_text,fmt='',cmap=sns.color_palette('magma'),linewidths=10,
            linecolor='lightgreen')
```

```python
df = sns.load_dataset('tips')
```

```python
# sns.lmplot
sns.lmplot(df,x = 'total_bill',y='tip',hue='sex',col='day',row='time',markers='x',sharex=True,sharey=True)
plt.show()
```

```python
h = sns.FacetGrid(data=df,row='smoker',col='day')
h.map_dataframe(sns.countplot,x='sex')
plt.show()
```

```python
h = sns.FacetGrid(data=df,row='smoker',col='day',hue = 'sex',legend_out=True,height=5, aspect=1)

h.map_dataframe(sns.scatterplot,x = 'tip',y = 'total_bill')
h.add_legend()
plt.show()
```

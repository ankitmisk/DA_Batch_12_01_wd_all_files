```python
import matplotlib.pyplot as plt
```

```python
df['species'].unique()
```

```python
dict_data = dict(zip(df['species'].unique(),range(3)))
```

```python
df['species'].apply(lambda x: dict_data[x])
```

```python
x = df['sepal_length']
y = df['sepal_width']

plt.scatter(x,y,c =df['species'].apply(lambda x: dict_data[x]))
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
x = df['petal_length']
y = df['petal_width']

plt.scatter(x,y,c =df['species'].apply(lambda x: dict_data[x]))
plt.xlabel('Petal Length')
plt.ylabel('Pepal Width')
plt.show()
```

```python
sns.scatterplot(data = df, x = x, y = y,hue = 'species')
```

```python
sns.barplot(data = tips, x='day',y = 'total_bill' ,
            errorbar=('ci', 95),hue = 'time',orient='v',estimator='max')
```

```python
ax = sns.barplot(data = tips, x='day',y = 'total_bill' ,
            errorbar=('ci', 0),hue = 'time')

for i in ax.containers:
    plt.bar_label(i,padding=4,rotation = 90)
    
plt.yticks(range(0,40,5))
plt.show()
```

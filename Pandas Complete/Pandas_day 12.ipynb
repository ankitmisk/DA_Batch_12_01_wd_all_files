```python
data1 = {'eid':np.arange(1,11),
        'Salary':np.random.randint(30000,70000,10)}

a = np.arange(1,11)
random.shuffle(a)

data2 = {'eid':a,
        'Sales':np.random.randint(0,1000000,10)}


df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
df2
```

```python
df1.join(df2.set_index('eid'),on = 'eid')
```

```python
df1.set_index('eid').join(df2.set_index('eid'),how = 'left')
```

```python
df1.set_index('eid').join(df2.set_index('eid'),how = 'right')
```

```python
data1 = {'eid':np.arange(5,15),
        'Salary':np.random.randint(30000,70000,10)}

a = np.arange(1,11)
random.shuffle(a)

data2 = {'eid':a,
        'Sales':np.random.randint(0,1000000,10)}


df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
import numpy as np
import random
```

```python
data1 = {'sid':np.arange(1,11),
         'Sci_marks':np.random.randint(1,100,10),
        'Maths_marks':np.random.randint(1,100,10),
        'Hindi_marks':np.random.randint(1,100,10),
        'common':np.random.randint(1,10,10)}

data2 = {'sid':np.arange(5,15),
         'Swimming_marks':np.random.randint(1,100,10),
        'Sports_marks':np.random.randint(1,100,10),
        'Music_marks':np.random.randint(1,100,10),
        'common':np.random.randint(1,10,10)}
```

```python
subject_score = pd.DataFrame(data1,index = list(range(5,15)))
extra_score = pd.DataFrame(data2)
```

```python
len(range(5,15))
```

```python
extra_score.merge(subject_score)
```

```python
# {'left', 'right', 'outer', 'inner', 'cross'}
```

```python
extra_score.merge(subject_score,how = 'left')
```

```python
extra_score.merge(subject_score,how = 'right')
```

```python
extra_score.merge(subject_score,how = 'outer')
```

```python
extra_score.merge(subject_score,how = 'outer',indicator=True)
```

```python
extra_score.merge(subject_score,how = 'outer',indicator=True,sort = True)
```

```python
extra_score.merge(subject_score,how = 'outer',indicator=True,on = 'sid',suffixes = ("_left","_right"))
```

```python
df_1949 = flights_df[flights_df['year'] == 1949]
df_1950 = flights_df[flights_df['year'] == 1950]
df_1951 = flights_df[flights_df['year'] == 1951]
```

```python
import matplotlib.pyplot as plt
```

```python
df_1949.plot(kind='line',x = 'month',y = 'passengers',color='r')
df_1950.plot(kind='line',x = 'month',y = 'passengers',color = 'g')
df_1951.plot(kind='line',x = 'month',y = 'passengers',color = 'b')
plt.show()
```

```python
plt.plot(df_1949['month'],df_1949['passengers'])
plt.plot(df_1950['month'],df_1950['passengers'])
plt.plot(df_1951['month'],df_1951['passengers'])

plt.show()
```
